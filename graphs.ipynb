```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Types of Graphs



# 1.Plot-        line graph, between two parameters
# 2.bar-         bar graph of single parameter
# 3.hist-        histogram of a bunch of data i.e. frequency of single parameter
# 4.scatter-     dot graph between two parameter 
# 5.boxplot/box- statstical overview is form of box for a single parameter
#                mean,outliers,quantiles(25%,75%),IQR 
#                (Optional-categorical data)
# 6.stem-        more like bar graph,lines instead of base_repr
# 7.violinpot-   violin shaped statistical overview graph 
#                spreads of data (more like hist) 
#                single parameter (optional cat-data)       
```

```python
x = np.linspace(0,5,11)
x1 = np.arange(1,11,3)
```

```python
y=x**3
```

```python
plt.plot(x,y)
```

```python
plt.bar(x,y)
plt.title('bar graph')
plt.show()
```

```python
plt.hist(x)
plt.title('histogram')
plt.show()
```

```python
plt.scatter(x,y)
plt.title('scatter')
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.yticks(range(0,200,100))
plt.show()
```

```python
plt.boxplot(x)
plt.title('boxplot')
plt.show()
```
